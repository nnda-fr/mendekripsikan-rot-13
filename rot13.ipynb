```python
def rot13_transform(text):
    transformed_text = ""
    for char in text:
        if 'a' <= char <= 'z':
            transformed_text += chr(((ord(char) - ord('a') - 13 + 26) % 26) + ord('a'))
        elif 'A' <= char <= 'Z':
            transformed_text += chr(((ord(char) - ord('A') - 13 + 26) % 26) + ord('A'))
        else:
            transformed_text += char
    return transformed_text

# Contoh penggunaan untuk dekripsi:
encrypted_message_1 = "Uryyb, Jbeyq!"
decrypted_message_1 = rot13_transform(encrypted_message_1)
print(f"Pesan terenkripsi: {encrypted_message_1}")
print(f"Pesan terdekripsi: {decrypted_message_1}")

encrypted_message_2 = "hyne zryvatxne qv ngnf cntne."
decrypted_message_2 = rot13_transform(encrypted_message_2)
print(f"\nPesan terenkripsi: {encrypted_message_2}")
print(f"Pesan terdekripsi: {decrypted_message_2}")

# Contoh penggunaan untuk enkripsi:
original_message = "haloo aku ananda safrida."
encrypted_message_3 = rot13_transform(original_message)
print(f"\nPesan asli: {original_message}")
print(f"Pesan terenkripsi ROT13: {encrypted_message_3}")
```
